```python
import altair as alt # Load altair for charting
from vega_datasets import data
import pandas as pd # Load pandas for data processing

import persist_ext as PR # Import the extension
```

```python
cars = data.cars()
cars.head()
```

```python
PR.vis.interactive_table(cars)
```

```python
PR.vis.barchart(cars, x="Cylinders:O", y="count()")
```

```python
added_power = PR.df.get("added_power")
added_power
```

```python
power_grouped = PR.df.get("power_grouped", groupby="Power", aggregate={"Miles_per_Gallon": "mean", "Cylinders": "mean", "Displacement": "mean", "Horsepower": "mean", "Weight_in_lbs": "mean", "Acceleration": "mean"})
power_grouped
```

```python
data = added_power

selection = alt.selection_interval("selector", encodings=["x", "y"])

scatterplot = alt.Chart(data).mark_point().encode(x="Miles_per_Gallon:Q", y="Weight_in_lbs:Q", color="Power:N").add_params(selection)

barchart = alt.Chart(data).mark_bar().encode(y="Power:N", x="count()", color="Power:N").transform_filter(selection)

scatterplot & barchart
```
