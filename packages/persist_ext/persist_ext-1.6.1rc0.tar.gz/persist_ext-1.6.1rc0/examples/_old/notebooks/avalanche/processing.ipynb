```python
import pandas as pd
```

```python
df_o = pd.read_csv("./avalanches_cleaned_.csv")
df = df_o.copy(deep=True)
df.head()
```

```python
df["D"] = df[";Date"]
df.head()
```

```python
def fn(d):
    d_parts = d.split("-")
    d = f"{d_parts[1]}-{d_parts[2]}-{d_parts[0]}"
    return d

df.D = df[";Date"].apply(fn)
df[[";Date", "D"]].head()
```

```python
df.convert_dtypes().head()
```

```python
df.D = pd.to_datetime(df.D)
df.dtypes
```

```python
df.D = df.D.apply(lambda x: f"{x.month}/{x.day}/{x.year}")
df
```

```python
df_o[";Date"] = df.D
df_o = df_o.rename(columns={";Date": "Date"})
df_o.head()
```

```python
def is_int_or_float(v):
    if isinstance(v, int) or isinstance(v, float):
        return True
    if isinstance(v, str):
        try:
            if "." in v:
                float(v)
            else:
                int(v)
            return True
        except:
            return False
    else:
        return False
        

df_o["Depth_inches"] = df_o["Depth_inches"].astype(str)
# df_o.loc[[0], ["Depth_inches"]] = "A"
df_o = df_o.convert_dtypes()
```

```python
for c in df_o.select_dtypes(include=["string"]):
    print(c)
```

```python
df_col_renamed = df_o.rename(columns={"Aspect": ";Aspect"})
```

```python
df_col_renamed.to_csv("./avalanches_cleaned.csv", index=False)
```
